```python
import subprocess
```

```python
def html_table_to_rst(html_string):
    try:
        process = subprocess.Popen(
            ['pandoc', '-f', 'html', '-t', 'rst', '--wrap=none'],
            stdin=subprocess.PIPE,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        
        rst_output, errors = process.communicate(input=html_string)
        
        if process.returncode != 0:
            raise Exception(f"Pandoc conversion failed: {errors}")
            
        return rst_output.strip()
        
    except Exception as e:
        raise Exception(f"Conversion failed: {str(e)}")
```

```python
def html_table_to_rst(html_string, first_row_is_header=True):
    try:
        table_data = html2data(html_string)
        
        if not table_data:
            return ""

        # print(table_data[0])
        table_data = [list(r) for r in table_data[0]]
        
        if first_row_is_header:
            header_length = len(table_data[0])
            table_data = [table_data[0], [""] * header_length] + table_data[1:]
            
        return data2rst(table_data)
    except Exception as e:
        raise Exception(f"Conversion failed: {str(e)}")

print(html_table_to_rst(html))
```
